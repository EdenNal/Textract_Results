```python
# I want a python function that opens a csv and counts the number of rows in it.
import csv
def count_csv_rows(file_path):
    with open(file_path, mode='r', newline='') as csvfile:
        csvreader = csv.reader(csvfile)
        row_count = sum(1 for row in csvreader)
    return row_count
print(count_csv_rows('TablesAndConfidence/cdi_ca_1956_wk_prov_dbs_Part32_tables.csv'))
```

```python
import csv

def remove_every_other_row(input_csv, output_csv, keep_header=True):
   
    with open(input_csv, newline='', encoding='utf-8') as infile, \
         open(output_csv, 'w', newline='', encoding='utf-8') as outfile:

        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        if keep_header:
            header = next(reader)
            writer.writerow(header)

        for i, row in enumerate(reader):
            if i % 2 == 0:   # keep rows 0,2,4,...
                writer.writerow(row)
        print(f"Processed {input_csv} and saved to {output_csv}")
for i in range (2, 36):
    remove_every_other_row(f'by disease/{i}.csv', f'by disease/{i}_filtered.csv')


```

```python
import csv
import re

def reorder_rows_remove_padded(input_csv, output_csv, keep_header=True):

    with open(input_csv, newline='', encoding='utf-8') as f:
        reader = csv.reader(f)
        rows = list(reader)

    if not rows:
        print("File empty.")
        return

    header = None
    data_rows = rows

    if keep_header:
        header = rows[0]
        data_rows = rows[1:]

    # Remove rows starting with "padded_"
    data_rows = [r for r in data_rows if not r[0].startswith("padded_")]

    part_re = re.compile(r'Part(\d+)', re.IGNORECASE)

    def extract_part(row):
        m = part_re.search(row[0])
        return int(m.group(1)) if m else 10**9

    # Sort numerically by Part#
    data_rows_sorted = sorted(data_rows, key=extract_part)

    with open(output_csv, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        if keep_header:
            writer.writerow(header)
        writer.writerows(data_rows_sorted)

    print("Reordered file written:", output_csv)

for i in range(2,36):
     reorder_rows_remove_padded(f'by disease/{i}_filtered.csv', f'by disease/{i}_reordered.csv')
```

```python
# python function that opens a csv and counts the number of rows in it until it finds a row that begins with a specific value, then stops and returns the count up to that point.
import csv
def count_csv_rows_until_value(file_path, stop_value):  
    with open(file_path, mode='r', newline='') as csvfile:
        csvreader = csv.reader(csvfile)
        row_count = 0
        for row in csvreader:
            if stop_value == row[0]:
                break
            row_count += 1
    return row_count
# print("we want to remove " + str(count_csv_rows_until_value('TablesAndConfidence/cdi_ca_1956_wk_prov_dbs_Part1_confidences.csv', '1'))+ " rows")

# Week 32 doesn't have a '1' value, so this will return the full count of rows!!!!!!!!!!!!!!!!! Will replace with 'K'
print(count_csv_rows_until_value('TablesAndConfidence/cdi_ca_1956_wk_prov_dbs_Part32_tables.csv', 'K'))
```

```python
# python function to remove empty rows from a csv file 
import csv
def remove_empty_rows(input_path, output_path):
    with open(input_path, "r", newline="", encoding="utf-8") as infile, \
         open(output_path, "w", newline="", encoding="utf-8") as outfile:

        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        for row in reader:
            # Check if all columns are empty after stripping whitespace
            if any(cell.strip() for cell in row):
                writer.writerow(row)



os.makedirs('Cleaned_Manual_CSVs', exist_ok=True)    
remove_empty_rows('1,2,3,4.csv', 'Cleaned_Manual_CSVs/1,2,3,4.csv')
# for i in range(1,53):
#     input_file = remove_empty_rows(f'cdi_ca_1956_wk_prov_dbs_manual/cdi_ca_1956_wk_prov_dbs_Part{i}.csv')
    
```

```python
import csv
import os

def pad_csv_to_padded_folder(infile, out_dir="", placeholder="FILLER"):
    

    # Ensure output directory exists
    os.makedirs(out_dir, exist_ok=True)

    rows = []
    max_cols = 0

    # First pass: read all rows and find max number of columns
    with open(infile, newline="", encoding="utf-8") as f:
        reader = csv.reader(f)
        for row in reader:
            rows.append(row)
            if len(row) > max_cols:
                max_cols = len(row)

    outfile = os.path.join(out_dir, "padded_" + os.path.basename(infile) )

    # Second pass: pad rows and write to new file
    with open(outfile, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        for row in rows:
            if len(row) < max_cols:
                row = row + [placeholder] * (max_cols - len(row))
            writer.writerow(row)

    print(f"Padded file written to: {outfile}")
    return outfile
# for i in range(1,53):
#     # remove each file that starts with padded in Cleaned_AWS_CSVs and Cleaned_Manual_CSVs
#     try:
#         os.remove(f'Cleaned_AWS_CSVs/paddedcdi_ca_1956_wk_prov_dbs_Part{i}_tables.csv')
#     except FileNotFoundError:
#         pass
#     try:
#         os.remove(f'Cleaned_AWS_CSVs/paddedcdi_ca_1956_wk_prov_dbs_Part{i}_confidences.csv')
#     except FileNotFoundError:
#         pass
#     try:
#         os.remove(f'Cleaned_Manual_CSVs/paddedcdi_ca_1956_wk_prov_dbs_Part{i}.csv')
#     except FileNotFoundError:
#         pass

for i in range(1,53):
    pad_csv_to_padded_folder(f'Removed_Columns_AWS_CSVs/padded_cdi_ca_1956_wk_prov_dbs_Part{i}_tables.csv', out_dir="Removed_Columns_AWS_CSVs")
    pad_csv_to_padded_folder(f'Removed_Columns_AWS_CSVs/padded_cdi_ca_1956_wk_prov_dbs_Part{i}_confidences.csv', out_dir="Removed_Columns_AWS_CSVs")
    pad_csv_to_padded_folder(f'Removed_Columns_Manual_CSVs/padded_cdi_ca_1956_wk_prov_dbs_Part{i}.csv', out_dir="Removed_Columns_Manual_CSVs")
```

```python
import os
import glob
import pandas as pd

# ---- CONFIG ----
INPUT_FOLDER = "C:\\Users\\edeno\\Downloads\\eden\\Textract Pipeline Experiments\\Cleaned_With_Columns\\Cleaned_AWS_CSVs"
OUTPUT_FOLDER = os.path.join(INPUT_FOLDER, "by_row_index")

def find_first_csv_like_line(lines, min_commas=2, max_scan_lines=80):
    for i, line in enumerate(lines[:max_scan_lines]):
        if line.count(",") >= min_commas:
            return i
    return 0

def split_line(line):
    # simple split; keeps empty cells if there are ,, sequences
    return line.rstrip("\n\r").split(",")

def separate_by_row_index_raw():
    os.makedirs(OUTPUT_FOLDER, exist_ok=True)

    files = sorted(glob.glob(os.path.join(INPUT_FOLDER, "*.csv")))
    if not files:
        raise FileNotFoundError(f"No .csv files found in {INPUT_FOLDER}")

    buckets = {}      # row_index -> list of rows (each row is a list of cells)
    max_cols = 0      # track max columns across everything so we can pad

    for fpath in files:
        with open(fpath, "r", encoding="utf-8", errors="ignore") as f:
            lines = f.read().splitlines()

        start = find_first_csv_like_line(lines, min_commas=2)
        data_lines = [ln for ln in lines[start:] if ln.strip() != ""]  # drop blank lines

        for i, ln in enumerate(data_lines):
            cells = split_line(ln)
            # prepend filename so you know which source file the row came from
            cells = [os.path.basename(fpath)] + cells

            max_cols = max(max_cols, len(cells))
            buckets.setdefault(i, []).append(cells)

        print(f"Processed {os.path.basename(fpath)} (start_line={start}, rows={len(data_lines)})")

    # write one output file per row index
    for i in sorted(buckets.keys()):
        out_path = os.path.join(OUTPUT_FOLDER, f"{i+1}.csv")  # 1.csv, 2.csv, ...
        with open(out_path, "w", encoding="utf-8", newline="") as out:
            for row in buckets[i]:
                # pad to max_cols so rows have consistent width
                if len(row) < max_cols:
                    row = row + [""] * (max_cols - len(row))
                out.write(",".join(row) + "\n")

    print(f"Done. Wrote {len(buckets)} files to: {OUTPUT_FOLDER}")

# run it
separate_by_row_index_raw()
```
